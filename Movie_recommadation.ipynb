```python
import pandas as pd
```

```python
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
merged_df = pd.merge(
    movies,
    credits,
    left_on="id",
    right_on="movie_id",
    how="inner"
)
```

```python
print("Merged Dataset Shape:", merged_df.shape)

merged_df.head(1)
```

```python
merged_df.rename(columns={'title_x': 'title'}, inplace=True)
```

```python
merged_df = merged_df[
    [
        'id',
        'title',
        'overview',
        'genres',
        'keywords',
        'cast',
        'crew',
        'vote_average',
        'vote_count',
        'popularity'
    ]
]
```

```python
merged_df.head(1)
```

```python
import ast
```

```python
def convert(text):
    items = ast.literal_eval(text)
    return [item['name'] for item in items]
```

```python
merged_df['genres'] = merged_df['genres'].apply(convert)
merged_df['keywords'] = merged_df['keywords'].apply(convert)
```

```python
def convert_cast(text):
    items = ast.literal_eval(text)
    return [item['name'] for item in items[:3]]
```

```python
merged_df['cast'] = merged_df['cast'].apply(convert_cast)
```

```python
def fetch_director(text):
    items = ast.literal_eval(text)

    directors = []

    for item in items:
        if item['job'] == 'Director':
            directors.append(item['name'])

    return directors
```

```python
merged_df['crew'] = merged_df['crew'].apply(fetch_director)
```

```python
list_cols = ['genres', 'keywords', 'cast', 'crew']

for col in list_cols:
    merged_df[col] = merged_df[col].apply(tuple)

merged_df.drop_duplicates(inplace=True)
```

```python
merged_df['overview'] = merged_df['overview'].apply(lambda x: x.split())
```

```python
def collapse(lst):
    return [i.replace(" ", "") for i in lst]
```

```python
merged_df['genres'] = merged_df['genres'].apply(collapse)
merged_df['keywords'] = merged_df['keywords'].apply(collapse)
merged_df['cast'] = merged_df['cast'].apply(collapse)
merged_df['crew'] = merged_df['crew'].apply(collapse)
```

```python
merged_df['tags'] = (
    merged_df['overview'] +
    merged_df['genres'] +
    merged_df['keywords'] +
    merged_df['cast'] +
    merged_df['crew']
)
```

```python
new_df = merged_df[['id', 'title', 'tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    words = []

    for word in text.split():
        words.append(ps.stem(word))

    return " ".join(words)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie):

    # Find the movie index
    movie_index = new_df[new_df['title'] == movie].index[0]

    # Similarity scores
    distances = similarity[movie_index]

    # Sort by similarity (highest first)
    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]  # Skip the first one (the movie itself)

    # Print recommendations
    for movie in movies_list:
        print(new_df.iloc[movie[0]].title)
```

```python
recommend("Avatar")
```

```python
def recommend(movie):

    movie_index = new_df[new_df['title'] == movie].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    recommended_movies = []

    for item in movies_list:
        recommended_movies.append({
            "id": new_df.iloc[item[0]].id,
            "title": new_df.iloc[item[0]].title
        })

    return recommended_movies
```

```python
recommend("Avatar")
```
